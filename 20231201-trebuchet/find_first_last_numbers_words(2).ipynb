## Problem Description

Your calculation isn't quite right. It looks like some of the digits are actually spelled out with letters: one, two, three, four, five, six, seven, eight, and nine also count as valid "digits".

Equipped with this new information, you now need to find the real first and last digit on each line. For example:

two1nine
eightwothree
abcone2threexyz
xtwone3four
4nineeightseven2
zoneight234
7pqrstsixteen
In this example, the calibration values are 29, 83, 13, 24, 42, 14, and 76. Adding these together produces 281.

What is the sum of all of the calibration values?

Your puzzle answer was 54578.



```python
# Read each row of the input file into a list of strings
with open('input1.txt') as f:
    content = f.readlines()

content = [x.strip() for x in content]
```

```python
content
```

```python
# Map numbers spelled out to their digit equivalent
num_dict = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9}
```

```python
sum = 0
for line in content:
    for digits in [re.findall(r'(?=(\d|' + "|".join(num_dict.keys()) + r'))' , line)]:
        digits = [int(digit) if digit not in num_dict.keys() else num_dict[digit] for digit in digits]
        sum += digits[0]*10 + digits[-1]

sum
```
